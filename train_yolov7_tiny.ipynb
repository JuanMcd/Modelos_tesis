```python
# Creamos reparameterize.py en /content/yolov7/models/
texto = """
import torch
import argparse
import os
from experimental import attempt_load

def reparameterize_model(weights, save_dir):
    device = 'cpu'
    model = attempt_load(weights, map_location=device)

    # Reparameterize if possible
    for m in model.modules():
        if hasattr(m, 'repvgg_convert'):
            m.repvgg_convert()

    os.makedirs(save_dir, exist_ok=True)
    save_path = os.path.join(save_dir, 'reparameterized_model.pt')
    torch.save(model.state_dict(), save_path)
    print(f'Model reparameterized and saved to {save_path}')

if __name__ == "__main__":
    parser = argparse.ArgumentParser()
    parser.add_argument('--weights', type=str, required=True, help='path to weights file')
    parser.add_argument('--save-dir', type=str, default='./reparam_model', help='directory to save reparameterized model')
    opt = parser.parse_args()

    reparameterize_model(opt.weights, opt.save_dir)
"""

# Escribimos el archivo
with open('/content/yolov7/models/reparameterize.py', 'w') as f:
    f.write(texto)

print("Script reparameterize.py creado correctamente.")
```
